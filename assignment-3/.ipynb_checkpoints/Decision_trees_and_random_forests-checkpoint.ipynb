# Assignment: Decision Trees and Random Forests

## 0. Setting Up the Data

From UCI Machine Learning Repository we first fetched the [**Phishing Websites**](https://archive.ics.uci.edu/dataset/327/phishing+websites) data set. 

```python
from ucimlrepo import fetch_ucirepo 
  
# fetch dataset 
phishing_websites = fetch_ucirepo(id=327) 
  
# data (as pandas dataframes) 
X = phishing_websites.data.features 
y = phishing_websites.data.targets 
  
# metadata 
print(phishing_websites.metadata) 
  
# variable information 
print(phishing_websites.variables) 
```

## 1. Business Understanding

The goal is to asses whether a website can be classified as phishing or legitimate based on website features using **Decision Tree** and **Random Forest**.

Model performance is evaluated to determine if a reliable phishing prediction is feasible with the given data and models.

## 2. Data Understanding

```python
# Inspect the data
X.info()
X.describe()

# Display the targets
print("\nTarget value counts:\n" ,y.value_counts())

# Display unique values for each feature
print("\nUnique values for each feature:")
for num, col in enumerate(X.columns):
    unique_vals = sorted(int(x) for x in X[col].unique())
    print(f"{num} {col}: {unique_vals}")
```

The dataset contains 11,055 entries and 30 columns. Each row represents a website, and the columns represent different website-related features. The feature values are encoded as -1, 0, and 1, describing whether a website characteristic looks bad, suspicious, or normal.

## 3. Data Preparation

The dataset is complete and consistent. Features and Targets are separated. No additional preparation required.

## 4. Modeling

### Data splitting
The data is split into two sets. One for training and one for testing. 
* Training: 70%
* Testing: 30%

```python
from sklearn.model_selection import train_test_split

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    X, 
    y, 
    test_size=0.3, 
    random_state=42
)
```

### Training the Decision Tree

Create model with the best deemed parameters

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV

# Initialize the Decision Tree Classifier
clf = DecisionTreeClassifier(
    max_depth=None,
    min_samples_split=2,
    min_samples_leaf=1,
    criterion="entropy",
    random_state=42)
clf.fit(X_train, y_train)

# Hyperparameter tuning using GridSearchCV
params = {
    'max_depth': [3, 5, 7, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 5],
    'criterion': ['gini', 'entropy']
}

grid = GridSearchCV(DecisionTreeClassifier(random_state=42), param_grid=params, cv=5)
grid.fit(X_train, y_train)

print("Best parameters:", grid.best_params_)

```

Create a model with restrained depth

```python
clf_5 = DecisionTreeClassifier(
    max_depth=5,
    min_samples_split=2,
    min_samples_leaf=1,
    criterion="entropy",
    random_state=42)
clf_5.fit(X_train, y_train)

clf_3 = DecisionTreeClassifier(
    max_depth=3,
    min_samples_split=2,
    min_samples_leaf=1,
    criterion="entropy",
    random_state=42)
clf_3.fit(X_train, y_train)
```

## 5. Evaluation

Evaluating the unresrained model

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Evaluate the model
y_pred = clf.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

Evaluating the restrained models

```python
y_pred_5 = clf_5.predict(X_test)
y_pred_3 = clf_3.predict(X_test)
print("Accuracy of depth 5:", accuracy_score(y_test, y_pred_5))
print("Accuracy of depth 3:", accuracy_score(y_test, y_pred_3))

print("\nClassification Report for depth 5:\n", classification_report(y_test, y_pred_5))
print("\nClassification Report for depth 3:\n", classification_report(y_test, y_pred_3))

print("\nConfusion Matrix for depth 5:\n", confusion_matrix(y_test, y_pred_5))
print("\nConfusion Matrix for depth 3:\n", confusion_matrix(y_test, y_pred_3))
```

Tests performed with max depth of 3, 5 and 7. Statistics of max depth 7 are left out and only referenced in comparison to depth 3 and 5.  
Accuracy of the model increases with the increase in maximum depth reaching its best score with no set maximum depth.  
Maximum depth of 5 stands out as an outlier where recall for recognising legitimate sites lowers considerably.  
Maximum depth of 3 and 7 are more in-line with the performance and statistics of the optimal hyperparameters.  

While the accuracy of the model would be at its highest with no set maximum depth, this makes the logic of the decision unreadable and most likely results in overfitting for the model.  
For this reason maximum depth is set to 3.

### Visualising the Decision Tree

```python
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree

fig = plt.figure(figsize = (10, 10))
plot_tree(clf_3, feature_names = X.keys(), class_names = ['legitimate', 'phishing'])
plt.show()
```

Going through the decision tree depth-first search yields to following logic to follow.  
First it should be checked whether the ssl final state for the site completes. In the event that it does it can be considered legitimate.  
Following that the name used for the website in the url should be considered. If it that passes scrutiny as well the page can be declared legitimate.  
Should the website url seem suspicious the url prefix (e.g. www.) and suffix (e.g. .fi) should be checked. Their trustworthyness will help make the final judgement on the legitimatecy of the site.  

If the ssl final state for the site fails the website name in the url should be checked. If the name looks legitimate its prefix and suffix should once more be judged and this will lead to a conclusion about the site's legitimacy.  
In the event that the name of the site doesn't pass your judgement it should then be judged solely on the basis of being a phishing site and declared as such.

## 6. Deployment

For the purpose of recognising phishing sites the model is inadequate. The precision is not there for reasonable training parameters and at its highest precision and recall, achieved with no max depth, leaves the model with high likelyhood of being overfitted.  
However the model does show some potential at blocking or filtering phishing sites with max depth set at 5.  
With this parameter the model aggressively rules in favour of determining sites as phishing at the cost of noticeable number of legitimate sites, and in some circumstances this could be an acceptable trade-off to make.
